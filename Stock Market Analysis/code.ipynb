```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
stock_data = yf.download("AAPL", start = '2020-01-01', end = '2025-01-01')
print(stock_data.head())  
```

```python
plt.figure(figsize = (12, 6))
plt.plot(stock_data['Close'] , label = "closing price", color = "blue")
plt.title("Apple Stock Closing Price")
plt.xlabel("Date")
plt.ylabel("Price (in USD)")
plt.legend()
plt.grid()
plt.show()
```

```python
stock_data['50_MA'] = stock_data['Close'].rolling(window = 50, min_periods = 1).mean()
stock_data['200_MA'] = stock_data['Close'].rolling(window = 200, min_periods =1).mean()
plt.figure(figsize = (12, 6))
plt.plot(stock_data['Close'], label = "closing price", color = "b")
plt.plot(stock_data['50_MA'], label = "50 day moving average", color = 'r')
plt.plot(stock_data['200_MA'], label = "200 day moving average", color = 'g')
plt.title("Apple Stock Price with moving average")
plt.xlabel("Date")
plt.ylabel("Price (in USD)")
plt.legend()
plt.grid()
plt.show()
```

```python
stock_data['Daily_Returns'] = stock_data['Close'].pct_change() 
print(stock_data[['Close', 'Daily_Returns']].head())

#plot daily returns
plt.figure(figsize = (12, 6))
plt.plot(stock_data['Daily_Returns'], label = 'Daily Returns', color = 'purple')
plt.axhline(0, linestyle='--', color='black')
plt.title("Daily Returns of Apple")
plt.xlabel("Date")
plt.ylabel("Return")
plt.legend()
plt.grid()
plt.show()
```
